В данных рассматривается период с 7 октября по 3 ноября 2019 года, всего 4 недели.

Составим профили пользователей:

```python
# определим дату первого посещения каждым пользователем
df1 = data.groupby('user_id').agg({'event_time':'min'}).rename(columns={'event_time':'first_ts'})
# составим таблицу профилей пользователей
profiles = channel.merge(df1, on='user_id')
# для когортного анализа определяем дату и неделю первого посещения
profiles['dt'] = profiles['first_ts'].dt.date
profiles
```

Составим функцию для расчета удержания по когортам по различным признакам:

```python
def get_retention(profiles, data, dimensions=[]):
    
    # составляем таблицу с данными о посещениях
    result_raw = profiles.merge(
        data[['user_id', 'event_time']], on='user_id'
    )
    
    # добавляем столбец с количеством дней, прошедших с момента первого посещения (лайфтайм)
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
                             ).dt.days
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']
    
    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result['cohort_size'] = cohort_sizes
        return result
    
    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions)
    
    # возвращаем обе таблицы и сырые данные
    return result_grouped
```

```python
# составим функцию для построения кривой удержания пользователей
def plot_retention(retention):
    retention.drop(columns=['cohort_size', 0]).T.plot(grid=True, figsize=(15,5))
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.ylabel('Коэффициент удержания')
    plt.title('Удержание пользователей')
    plt.show()
```

```python
# построим таблицу удержания по всем пользователям
retention = get_retention(profiles, data)
```

```python
# построим таблицу удержания по дневным когортам
retention_dt = get_retention(
    profiles, data, dimensions='dt'
                            ).drop(index=dt.date(2019, 11, 3))
```

```python
# построим кривую удержания пользователей
sns.set_style('darkgrid')
plot_retention(retention)
```

```python
# построим тепловую карту удержания пользователей по дневным когортам
plt.figure(figsize=(20, 13))
sns.heatmap(retention_dt.drop(columns=['cohort_size', 0]), 
            annot=True, linewidth=1, fmt='.2f')
plt.xlabel('Лайфтайм')
plt.ylabel('Когорта')
plt.title('Коэффициент удержания по дневным когортам')
plt.show()
```

На графиках видно, что удержание пользователей у мобильного приложения довольно низкое с самого начала.\
Из всех пользователей в целом лишь 11% возвращаются на следующий день после своего первого события, через неделю - 5%, через 2 недели - 2%, через 3 недели - 1%.\
Если анализировать дневные когорты, то на следующий день вернулось больше всех пользователей, впервые зашедших 29 октября (17%), меньше всех - впервые зашедших 9 и 27 октября (7%). Данные по доступным для каждой из когорт последним дням (диагональ карты) колеблятся от 0 до 5%, что также не является высокими показателями. 

#### Анализ продолжительности сессий

##### Создание журнала сессий

Для определения продолжительности сессии создадим столбец с разницами во времени между событиями, сгрупированный по пользователям и по дням. Разницами между временем событий, являющиеся достаточно маленькими, чтобы их относить к одной сессии и при этом приходящиеся на разные дни, можно пренебречь, так как их несоизмеримо меньше, чем остальных разниц (максимальное количество таких разностей для группы пользователь-день равно 1). 

```python
# добавим в историю посещений столбец с датой посещения
data['dt'] = data['event_time'].dt.date
# отсортируем датасет по пользователию и времени посещения
data.sort_values(by=['user_id', 'event_time'], inplace=True)
# создадим столбец с разницами между событиями
data['diff'] = data.groupby(['user_id', 'dt'])['event_time'].diff()
# переведем столбец из timedelta в дату и время
data['diff'] = data['diff'].dt.seconds
```

```python
# построим диаграмму размаха для разницы между событиями
plt.figure(figsize=(12,7))
sns.boxplot(data=data, y='diff')
plt.ylabel('Разница в секундах')
plt.title('Диаграмма размаха разницы между событиями в секундах')
plt.show()
```

На графике видно, что между большинством событий в рамках суток промежуток очень маленький, остальные данные являются выбросами. Построим тот же самый график с ограничением на диапазон значений вертикальной оси, чтобы рассмотреть распределение данных.

```python
# построим диаграмму размаха с ограничением на вертикальную ось
plt.figure(figsize=(12,7))
sns.boxplot(data=data, y='diff')
plt.ylabel('Разница в секундах')
plt.title('Диаграмма размаха разницы между событиями в секундах')
plt.ylim((0,1800))
plt.show()
```

Максимальное значение верхнего уса составляет менее 400 секунд, данные выше этого порога являются выбросами. Поэтому в качестве значения, выше которого перерыв между событиями означает начало новой сессии, можно взять 420 секунд (7 минут).

```python
# разница между временем событий для каждого из пользователей
diff_timestamp = data.groupby('user_id')['event_time'].diff()
# индексы для каждой новой сессии
new_session = (diff_timestamp.isnull()) | (diff_timestamp > pd.Timedelta('7Min'))
# создаем номера уникальных сессий для каждого из пользователей
data['session_id'] = (
    data.loc[new_session, ['user_id', 'event_time']]
    .groupby('user_id').rank(method='first').astype(int)
)
# заполняем значения для событий в рамках одной сессии
data['session_id'] = data['session_id'].fillna(method='ffill').astype(int)
```

```python
# составим журнал сессий
sessions = data.groupby(['user_id','session_id']).agg({'event_time':['min','max']}).droplevel(1, axis=1)
sessions.columns=['start', 'end']
sessions.reset_index(inplace=True)
sessions['duration'] = sessions['end'] - sessions['start']
# переведем столбец duration из timedelta в дату и время в минутах
sessions['duration'] = sessions['duration'].dt.seconds/60
sessions
```

##### Расчет конверсии в просмотр контактов

Составим функцию для расчета конверсии по лайфтаймам для когорт по различным признакам:

```python
def get_conversion(profiles, data, dimensions=[]):  
    # определяем дату и время первого просмотра контактов для каждого пользователя
    contacts_df = (
        data[data['event_name']=='contacts_show'].sort_values(by=['user_id', 'event_time'])
        .groupby('user_id')
        .agg({'event_time': 'first'})
        .reset_index()
    )
    # добавляем данные о просмотре контактов в профили
    result_raw = profiles.merge(
        contacts_df, on='user_id', how='left')
    
    # рассчитываем лайфтайм для каждого просмотра контактов
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
    ).dt.days
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']
        
    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        result = result.fillna(0).cumsum(axis = 1)
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # делим каждую «ячейку» в строке на размер когорты
        # и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        result['cohort_size'] = cohort_sizes
        return result
    
    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions)
    
    return result_grouped
```

```python
# построим таблицу конверсии по всем пользователям
conversion = get_conversion(profiles, data)
```

```python
# построим кривую конверсии пользователей
def plot_conversion(conversion):
    conversion.drop(columns='cohort_size').T.plot(grid=True, figsize=(15,5))
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.ylabel('Коэффициент конверсии')
    plt.title('Конверсия пользователей')
    plt.show()
plot_conversion(conversion)
```

Исходя из проведенного анализа, в целом можно выделить всего 5 событий, из которых целесообразно пытаться выстраивать цепочки событий. Это photos_show, search_1, map, contacts_show, advert_open.

```python
# построим профили пользователей с масками событий
user_way = data[['user_id']]
for event in ['contacts_show', 'photos_show', 'search_1', 'map', 'advert_open']:
    user_way[event] = data['event_name']==event
user_way = user_way.groupby('user_id').agg('max')
user_way
```

```python
# составим функцию для создания датафреймов с воронками из 3 событий (сквозная и от предыдущего этапа) 
def df_funnel_3(event_1, event_2, event_3):
    a = len(user_way[user_way[event_1]==True])
    b = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_2]==True)]))
    c = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_2]==True) 
                      & (user_way[event_3]==True)]))     
    d = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_3]==True)]))
    df1 = pd.DataFrame([a, b, c], index=[event_1, event_2, event_3], columns=['user_cnt'])
    df1['cr, %'] = round(df1['user_cnt']/df1.loc[event_1, 'user_cnt']*100, 2)
    df1['cr_step, %'] = round((df1['user_cnt'].pct_change() + 1)*100, 2)
    df2 = pd.DataFrame([a, b, d], index=[event_1, event_2, event_3], columns=['user_cnt'])
    df2['cr, %'] = round(df2['user_cnt']/df2.loc[event_1, 'user_cnt']*100, 2)
    return df1, df2
```

```python
# составим функцию для создания датафреймов с воронками из 4 событий (сквозная и от предыдущего этапа) 
def df_funnel_4(event_1, event_2, event_3, event_4):
    a = len(user_way[user_way[event_1]==True])
    b = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_2]==True)]))
    c = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_2]==True) 
                      & (user_way[event_3]==True)]))
    d = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_2]==True) 
                      & (user_way[event_3]==True) 
                      & (user_way[event_4]==True)]))     
    e = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_3]==True)]))
    f = (len(user_way[(user_way[event_1]==True) 
                      & (user_way[event_4]==True)]))
    df1 = pd.DataFrame([a, b, c, d], index=[event_1, event_2, event_3, event_4], columns=['user_cnt'])
    df1['cr, %'] = round(df1['user_cnt']/df1.loc[event_1, 'user_cnt']*100, 2)
    df1['cr_step, %'] = round((df1['user_cnt'].pct_change() + 1)*100, 2)
    df2 = pd.DataFrame([a, b, e, f], index=[event_1, event_2, event_3, event_4], columns=['user_cnt'])
    df2['cr, %'] = round(df2['user_cnt']/df2.loc[event_1, 'user_cnt']*100, 2)
    return df1, df2
```

Лучше в качестве признака сегментации рассмотреть частоту действий, так как этот признак задан в исходных данных. Из-за высокой корреляции с длительностью сессии по нему заодно можно судить и о характере влияния продолжительности сессии на коэффициент удержания и конверсии.

В случае с частотой действий причина разницы может заключаться в том, что пользователи из сегментов с худшими показателями совершают не те события, которые совершаются пользователями из сегментов с лучшими показателями. Если разница в конверсии и удержании будет выявлена, то будет иметь смысл сосредоточить внимание на сравнении пользовательских путей отдельных сегментов, а не всех пользователей сразу.\
Предполагается, что у пользователей с более высокой частотой более высокие конверсия и удержание.

Разделим количество пользователей на 4 сегмента по квартилям частоты действий.

```python
# построим функцию для сегментации
def categorize_count(event_count):    
    if event_count<=profiles['event_count'].quantile(0.25):
        return 'Низкая частота'
    elif profiles['event_count'].quantile(0.25)<event_count<=profiles['event_count'].quantile(0.5):
        return 'Средняя частота'
    elif profiles['event_count'].quantile(0.5)<event_count<=profiles['event_count'].quantile(0.75):
        return 'Высокая частота'
    else:
        return 'Очень высокая частота'

# применим функцию для создания нового столбца с названием сегмента
profiles['segment'] = profiles['event_count'].apply(categorize_count)
```

### Проверка статистических гипотез по источнику установки приложения

#### Выбор уровня значимости с учетом множественности проверок

Для того чтобы выбрать правильный уровень значимости с учетом множественности проверок, необходимо определиться с количеством проверяемых на одних и тех же данных гипотез.\
Всего будем проверять 5 гипотез: о различии в коэффициентах удержания на начало и конец лайфтайма, о различии в коэффициентах конверсии, о различии в средних продолжительностях сессии, о различии в средних частотах действий.

Поправка Бонферрони силнее всего снижает мощность теста.\
Поправка Холма и иные пошаговые методы являются наиболее подходящими с точки зрения мощности теста методами, однако в силу более сложного расчета уровня значимости и необходимости ранжирования гипотез по возрастанию p-value использование данного метода оправдано лишь в случае спорных значений p-value. \
Поэтому для первого тестирования всех гипотез лушче всего использовать поправку Шидака, так как она меньше Бонферрони снижает мощность теста, а уровень значимости одинаков для всех проверок в отличие от метода Холма. Возьмем базовый уровень значимости равным 0.05. Тогда с учетом поправки и количества гипотез уровень значимости получается равным следующему числу:

```python
alpha = 0.05
m = 5
alpha_new = 1-(1-alpha)**(1/m)
print('Уровень значимости:', round(alpha_new,4))
```

#### Сравнение коэффициентов удержания

```python
# построим функцию для проведения z-теста для долей
def z_testing(df, success):    
    p_value = (
        pr.proportions_ztest([df.loc['google', success], df.loc['yandex', success]], 
                             [df.loc['google', 'cohort_size'], df.loc['yandex', 'cohort_size']])[1]
    )
    print('p-значение:', round(p_value,5))
    if p_value < alpha_new:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница.')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными.')  
```

```python
# построим таблицу удержания для сегментов
retention_source = get_retention(profiles, data, 'source')
retention_source
```

```python
# подготовим датафрейм для z-теста для долей
df_ret = retention_source[['cohort_size']].drop(index='other')
df_ret['ret_init'] = df_ret['cohort_size'] * retention_source[1]
df_ret['ret_init'] = df_ret['ret_init'].astype('int')
df_ret['ret_fin'] = df_ret['cohort_size'] * retention_source[26]
df_ret['ret_fin'] = df_ret['ret_fin'].astype('int')
df_ret
```

Сформулируем гипотезы:\
H0: Доли повторно зашедших на следующий день после 1-ой сессии в выборках по источникам google и yandex равны.\
H1: Доли повторно зашедших на следующий день после 1-ой сессии в выборках по источникам google и yandex разные.

```python
# проверим гипотезу об удержании в 1-й лафтайм 
z_testing(df_ret, 'ret_init')
```

Сформулируем гипотезы:\
H0: Доли повторно зашедших в последний лайфтайм в выборках по источникам google и yandex равны.\
H1: Доли повторно зашедших в последний лайфтайм в выборках по источникам google и yandex разные.

```python
# проверим гипотезу об удержании в последний лафтайм 
z_testing(df_ret, 'ret_fin')
```

#### Сравнение средней продолжительности сессии

Сформулируем гипотезы:\
H0: Средняя продолжительность сессии в источниках google и yandex одинаковая.\
H1: Средняя продолжительность сессии в источниках google и yandex разная.

Проверим нормальность распределения данных для продолжительности сессии с помощью гистограмм и теста Шапиро-Уилка.

```python
ax = (
    profiles[profiles['source']=='google']['mean_duration']
    .plot(kind='hist', figsize=(20,5), ax=plt.subplot(1,2,1))
)
ax.set_xlabel('Средняя продолжительность сессии')
ax.set_ylabel('Количество пользователей')
ax.set_title('Гистограмма распределения средней продолжительности сессии google')
ax1 = (
    profiles[profiles['source']=='yandex']['mean_duration']
    .plot(kind='hist', figsize=(20,5), ax=plt.subplot(1,2,2))
)
ax1.set_xlabel('Средняя продолжительность сессии')
ax1.set_ylabel('Количество пользователей')
ax1.set_title('Гистограмма распределения средней продолжительности сессии yandex')
plt.show()
```

```python
print('google:', st.shapiro(profiles[profiles['source']=='google']['mean_duration']))
print('yandex:', st.shapiro(profiles[profiles['source']=='yandex']['mean_duration']))
```

Обе контрольные группы распределены ненормально. На гистограмме четко видно, что чем меньше средняя продолжительность сессии, тем больше пользователей в каждой из выборок. Из теста Шапиро-Уилка также следует вывод, что гипотезу о нормальности отклоняем (p-value < 0.0102 для каждой выборки).\
Таким образом, для проверки гипотезы о равенстве средних следует применять непараметрический тест Манна-Уитни.

```python
# построим функцию для теста Манна-Уитни
def mann_tester(measure, name):  
    p_value = st.mannwhitneyu(profiles[profiles['source']=='google'][measure],
                        profiles[profiles['source']=='yandex'][measure])[1]
    
    print('p-значение:', round(p_value,5))
    if p_value < alpha_new:
        print('Отвергаем нулевую гипотезу: между средними есть значимая разница.')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать средние разными.')  
    
    print(name, 'google:', 
         round(profiles[profiles['source']=='google'][measure].mean(), 2))
    print(name, 'yandex:', 
         round(profiles[profiles['source']=='yandex'][measure].mean(), 2))
    
    print('Относительная разница средней google к yandex:', 
          '{0:.1f}%'.format((profiles[profiles['source']=='google'][measure].mean()
                             /profiles[profiles['source']=='yandex'][measure].mean()-1)*100))
```

```python
mann_tester('mean_duration', 'Средняя продолжительность сессии')
```

Относительная разница в 8,8% между средними продолжительностями сессий google и yandex составляет лишь 0,69 минуты в абсолютных значениях.\
Тест Манна-Уитни также подтвердил отсутствие различий между двумя средними. Однако стоит отметить низкое значение p-value, например без поправки множественности результат был бы иным.\

Рекомендации:
* Сегментация пользователей показала: чем выше частота действий, тем выше конверсия и удержание. Таким образом, необходимо более тщательно рассмотреть поведение пользователей с очень высокой частотой и сравнить его с другими сегментами. Возможно, дело в том, что пользователи с меньшей частотой действий начинают не с тех событий, что пользователи с более высокой частотой.
* Необходимо улучшить формирование данных о пользовательских сессиях: добавить в обработку данных флаги 'начало сессии' и 'конец сессии', что позволит точно фиксировать продолжительность и количество сессий. Или принять стандартную продолжительность сессии за определенный промежуток времени, например 5 минут. Но тогда в данных могут быть неточности.
